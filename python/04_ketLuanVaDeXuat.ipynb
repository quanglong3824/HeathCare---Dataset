### 3.2 Táº¡o risk score framework

```python
# Táº¡o framework Ä‘Ã¡nh giÃ¡ nguy cÆ¡ dá»±a trÃªn insights
def tinhDiemNguyCo(tuoi, gioiTinh, benhTim, tangHuyetAp, glucose, bmi, hut_thuoc):
    """
    TÃ­nh Ä‘iá»ƒm nguy cÆ¡ Ä‘á»™t quá»µ dá»±a trÃªn cÃ¡c yáº¿u tá»‘
    """
    diem = 0
    
    # Tuá»•i tÃ¡c (0-4 Ä‘iá»ƒm)
    if tuoi >= 80:
        diem += 4
    elif tuoi >= 70:
        diem += 3
    elif tuoi >= 60:
        diem += 2
    elif tuoi >= 50:
        diem += 1
    
    # Giá»›i tÃ­nh (0-1 Ä‘iá»ƒm)
    if gioiTinh.lower() == 'male':
        diem += 1
    
    # Bá»‡nh tim (0-3 Ä‘iá»ƒm) - Yáº¿u tá»‘ quan trá»ng nháº¥t
    if benhTim:
        diem += 3
    
    # TÄƒng huyáº¿t Ã¡p (0-2 Ä‘iá»ƒm)
    if tangHuyetAp:
        diem += 2
    
    # Glucose (0-2 Ä‘iá»ƒm)
    if glucose > 200:
        diem += 2
    elif glucose > 140:
        diem += 1
    
    # BMI (0-1 Ä‘iá»ƒm)
    if bmi > 30:
        diem += 1
    
    # HÃºt thuá»‘c (0-1 Ä‘iá»ƒm)
    if hut_thuoc:
        diem += 1
    
    return diem

def danhGiaNguyCo(diem):
    """
    ÄÃ¡nh giÃ¡ má»©c Ä‘á»™ nguy cÆ¡ dá»±a trÃªn Ä‘iá»ƒm sá»‘
    """
    if diem >= 10:
        return "Ráº¤T CAO", "Cáº§n can thiá»‡p y táº¿ ngay láº­p tá»©c"
    elif diem >= 7:
        return "CAO", "Cáº§n theo dÃµi cháº·t cháº½ vÃ  Ä‘iá»u trá»‹ tÃ­ch cá»±c"
    elif diem >= 4:
        return "TRUNG BÃŒNH", "Cáº§n thay Ä‘á»•i lá»‘i sá»‘ng vÃ  theo dÃµi Ä‘á»‹nh ká»³"
    elif diem >= 2:
        return "THáº¤P", "Duy trÃ¬ lá»‘i sá»‘ng lÃ nh máº¡nh"
    else:
        return "Ráº¤T THáº¤P", "Tiáº¿p tá»¥c duy trÃ¬ tÃ¬nh tráº¡ng hiá»‡n táº¡i"

# VÃ­ dá»¥ sá»­ dá»¥ng
print("ğŸ¥ FRAMEWORK ÄÃNH GIÃ NGUY CÆ  Äá»˜T QUá»´")
print("=" * 50)

# Test cases
test_cases = [
    (45, 'Female', False, False, 90, 25, False),  # TrÆ°á»ng há»£p nguy cÆ¡ tháº¥p
    (65, 'Male', True, True, 180, 32, True),      # TrÆ°á»ng há»£p nguy cÆ¡ cao
    (75, 'Male', True, True, 220, 35, True),      # TrÆ°á»ng há»£p nguy cÆ¡ ráº¥t cao
]

for i, (tuoi, gioi, tim, ha, glucose, bmi, thuoc) in enumerate(test_cases, 1):
    diem = tinhDiemNguyCo(tuoi, gioi, tim, ha, glucose, bmi, thuoc)
    muc_do, khuyenNghi = danhGiaNguyCo(diem)
    
    print(f"\nğŸ“‹ TRÆ¯á»œNG Há»¢P {i}:")
    print(f"   ThÃ´ng tin: {tuoi} tuá»•i, {gioi}, Tim: {tim}, HA: {ha}, Glucose: {glucose}, BMI: {bmi}, Thuá»‘c: {thuoc}")
    print(f"   Äiá»ƒm nguy cÆ¡: {diem}/14")
    print(f"   Má»©c Ä‘á»™: {muc_do}")
    print(f"   Khuyáº¿n nghá»‹: {khuyenNghi}")
```
